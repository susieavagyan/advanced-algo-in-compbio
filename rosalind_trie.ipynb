```python
with open("rosalind_trie (1).txt") as f:
    strings = [line.rstrip() for line in f.readlines()]
```

```python
def build_trie(strings):
    trie = {1: {}}
    node_count = 1
    for string in strings:
        current_node = 1
        for i in range(len(string)):
            current_symbol = string[i]
            if current_symbol in trie[current_node]:
                current_node = trie[current_node][current_symbol]
            else:
                node_count += 1
                trie[current_node][current_symbol] = node_count
                current_node = node_count
                trie[current_node] = {}
    return trie

def print_trie(trie):
    for parent, children in trie.items():
        for base, child in children.items():
            print(parent, child, base)

# Example usage:

trie = build_trie(strings)
print_trie(trie)
```
