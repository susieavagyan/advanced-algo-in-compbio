```python
with open("debrujin.txt") as f:
    strings = [line.rstrip() for line in f.readlines()]

```

```python
k = len(strings[1])-1
```

```python

def reverse_complement(kmer):
    complement = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'}
    return ''.join([complement[base] for base in kmer[::-1]])


def de_bruijn_graph(strings, k):
    # Create a set of all (k+1)-mers and their reverse complements
    kp1mers = set()
    for string in strings:
            kp1mers.add(string)
            kp1mers.add(reverse_complement(string))

    # Create the graph as an adjacency list
    graph = set()
    for kmer in kp1mers:
        graph.add((kmer[:k],kmer[1:]))

    for edge in graph:
         print(edge)




```

```python
de_bruijn_graph(strings,k)
```
