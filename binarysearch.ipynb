```python
with open("rosalind_bins.txt") as f:
    strings = [line.rstrip() for line in f.readlines()]
```

```python
n,m, A, L = int(strings[0]), int(strings[1]), [ int(x) for x in strings[2].split() ], [ int(x) for x in strings[3].split() ]
```

```python
n,m
```

```python
A = sorted(A)
```

```python
def binary_search(A, key):
    left = 0
    right = n - 1
    while left <= right:
        mid = (left + right) // 2
        if A[mid] == key:
            return mid
        elif A[mid] < key:
            left = mid + 1
        else:
            right = mid - 1
    return -1

idx = []
for i in L:
    index = binary_search(A, i)
    if index != -1:
        index = index + 1
    idx.append(index)


print(' '.join(map(str, idx)))
```

```python
def binary_search2(A, key):
    left = -1 #rightmost element for which A[left] <= key
    right = n #leftmost element for which A[right] > key
    while right - left - 1 > 0:
        mid = (left + right) // 2
        if A[mid] == key:
            return mid
        if A[mid] > key:
            right = mid
        else:
            left = mid
    return -1

idx = []
for i in L:
    index = binary_search2(A, i)
    if index != -1:
        index = index + 1
    idx.append(index)


print(' '.join(map(str, idx)))
    
```
